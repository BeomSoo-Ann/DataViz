### Question 1) Regular plots [5 points]

Starting from the code below, try to reproduce the graphic below taking care of filled areas.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 256
X = np.linspace(-np.pi,np.pi,n,endpoint=True)
Y = np.sin(2*X)

plt.plot (X, Y+1, color='blue', alpha=1.00)
plt.plot (X, Y-1, color='blue', alpha=1.00)
plt.show()
```

<img src="regular_plots.png">

```python
# YOUR CODE HERE

fig, ax = plt.subplots()
ax.plot(X, Y+1, color='blue', alpha=1.00)
ax.plot(X, Y-1, color='blue', alpha=1.00)
ax.fill_between(X, Y+1, 1, color='blue', alpha=1.00)
ax.fill_between(X, Y-1, -1, color='blue', where=(Y-1 >= -1), alpha=1.00)
ax.fill_between(X, Y-1, -1, color='red', where=(Y-1 < -1), alpha=1.00)


plt.show()
```

### Question 4) Contour plots [5 points]

Starting from the code below, try to reproduce the graphic below taking care of the colormap.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)
X,Y = np.meshgrid(x,y)

plt.contourf(X, Y, f(X,Y), 8, alpha=.75, cmap='jet')
C = plt.contour(X, Y, f(X,Y), 8, colors='black', linewidth=.5)
plt.show()
```

### Question 13) Plot Job Year [10 pts]

Plot the line graph between number of job postings with respect to year. Print the year and the number of job postings as integer value. Note that year should be in ascending order.

```python
# YOUR CODE HERE
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings(action='ignore')

amazon_jobs = pd.read_csv('amazon_jobs_dataset.csv')

# extract only year from date
amazon_jobs['year'] = amazon_jobs['Posting_date'].str.split(',').str[1]

jobs_count_by_year = amazon_jobs['year'].value_counts().sort_index()

print(jobs_count_by_year)


fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.lineplot(x=jobs_count_by_year.index, y=jobs_count_by_year.values, palette="rocket")


plt.show()
```

### Question 14) Plot Month Job [10 pts]
 
Plot the bar graph between Month vs Job Openings. Print the month and the number of job posting as integer value. Note that the order of months doesn't matter.  

```python
# YOUR CODE HERE

# extract only month from date
amazon_jobs['month'] = amazon_jobs['Posting_date'].str.split(' ').str[0]
jobs_count_by_month = amazon_jobs['month'].value_counts().sort_index()

print(jobs_count_by_month)

fig, ax = plt.subplots(figsize=(10, 5))
ax = sns.barplot(x=jobs_count_by_month.index, y=jobs_count_by_month.values, palette="rocket")
ax.bar_label(ax.containers[0], label_type='edge', fontsize=12)

plt.show()
```

### Question 15) Plot City Job [10 pts]

Plot the pie chart between Indian cities vs number of jobs opening. Print the Indian cities and %age of Job distribution in India upto 2 decimal places. Note that %age of Job distribution should be in descending order.

```python
# YOUR CODE HERE

# extract country IN from location
amazon_jobs_IN = amazon_jobs[amazon_jobs['location'].str.contains('IN')]

# extract city from location
amazon_jobs_IN['city'] = amazon_jobs_IN['location'].str.split(',').str[-1]
jobs_count_by_INcity = amazon_jobs_IN['city'].value_counts().sort_index()

print(jobs_count_by_INcity)

# pie chart
fig, ax = plt.subplots(figsize=(10, 5))
ax.pie(jobs_count_by_INcity.values, labels=jobs_count_by_INcity.index, autopct='%.2f%%')


plt.show()

```
